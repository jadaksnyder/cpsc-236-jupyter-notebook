```python
print("Letter Grade Converter")

while True:
    numgrade = int(input("Enter numerical grade: "))

    if numgrade >= 88:
        lettergrade = "A"
    elif numgrade >= 80 and numgrade <= 87:
        lettergrade = "B"
    elif numgrade >= 67 and numgrade <= 79:
        lettergrade = "C"
    elif numgrade >= 60 and numgrade <= 66:
        lettergrade = "D"
    else:
        lettergrade = "F"

    print(f"Letter grade: {lettergrade}")

    user_input = input("Would you like to continue? (y/n): ").lower()

    if user_input != "y":
        print("Bye!")
        break
```

```python
print("Change Calculator")

while True:
    cents = int(input("Enter number of cents (0-99): "))
    quarters = cents // 25
    dimes = (cents % 25) // 10
    nickels = ((cents % 25) % 10) // 5
    pennies = ((cents % 25) % 10) % 5

    print(f"Quarters: {quarters}")
    print(f"Dimes: {dimes}")
    print(f"Nickels: {nickels}")
    print(f"Pennies: {pennies}")

    user_input = input("Would you like to continue? (y/n): ").lower()

    if user_input != "y":
        print("Bye!")
        break
```

```python
print("Table of Powers")

while True:
    startnum = int(input("Start number: "))
    stopnum = int(input("Stop number: "))
    
    if startnum < stopnum:
        break
    else:
        print("Please enter a start number that is less than the stop number.")

print("\nTable of Powers")
print("Number\tSquared\tCubed")
print("======\t=======\t=====")

for number in range(startnum, stopnum + 1):
    square = number ** 2
    cube = number ** 3
    print(f"{number}\t{square}\t{cube}")
```
